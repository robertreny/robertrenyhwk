```python
def fahr_to_celsius(temp):
    return((temp - 32) * (5/9))

#made function 
```

```python
fahr_to_celsius(40)
```

```python
print('freezing point of water: ', fahr_to_celsius(32), 'C')
```

```python
def celsius_to_kelvin(tempc):
    return(tempc + 273.15)
```

```python
celsius_to_kelvin(10)
```

```python
def far_to_kelvin(tempf):
    temp_k = celsius_to_kelvin(fahr_to_celsius(tempf))
    return(temp_k)
```

```python
far_to_kelvin(65)
```

```python
# visualize('infllamation-01.csv', ',')
# delim = ',' sets a default value if we dont provide the information 
def visualize(filename, delim=','):

#goal: create a function thfilenameat will visualize data easily:
#1 read in data file, assume numpy is in
    data = numpy.loadtxt(fname = filename, delimiter = delim)
#2 create our 3 subplots holders with the figure
    fig = matplotlib.pyplot.figure(figsize = (10.0, 3.0))
    axes1 = fig.add_subplot(1,3,1) #1 row 3 columns spec and place 1
    axes2 = fig.add_subplot(1,3,2)
    axes3 = fig.add_subplot(1,3,3)

#3 plot eahc of them 
    axes1.set_ylabel('average')
    axes1.plot(numpy.mean(data, axis = 0))

    axes2.set_ylabel('max')
    axes2.plot(numpy.max(data, axis = 0))

    axes3.set_ylabel('min')
    axes3.plot(numpy.min(data, axis = 0))
#4 we want to show plot
    fig.tight_layout()
    matplotlib.pyplot.show()






```

```python
import numpy 
import matplotlib.pyplot
```

```python
visualize('inflammation-01-Copy1.csv')
```

```python
#document function to remember what you did with a help function using ''''''
def offset_mean(data, target_mean_value = 0):
    '''Return a new arrayc = containing the original data with its mean
    offset to match the desired value, (0 by default)
    
    
    @param data an n-dimensional array of values
    @param target_mean_value the value you want to offset by (0 by default)
    @return a new array containing the original data with tis mean offset to match the desired value
    
    Examples
    ------------
    >>>> offset_mean([1,2,3],)
    '''
    return (data - numpy.mean(data)) + target_mean_value
```
