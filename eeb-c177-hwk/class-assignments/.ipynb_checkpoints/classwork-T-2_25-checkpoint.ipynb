```python
import csv
import re
```

```python
inputpath = "bee_list.txt"
```

```python
authors = []
with open(inputpath, "r") as dset:
    data = csv.DictReader(dset, delimiter = "\t")
    for dic in data:
        authors.append(dic["Taxon Author"])
    header = data.fieldnames # use this bc of jupyter I/O file close
```

```python
def check_auth(listofauthors):
    char = re.compile(r'.*Hirashima.*') # finish this with a regex pattern
    count = 0
    for author in listofauthors:
        l = char.search(author)
        d = (type(l) == re.Match)
        if d == True:
            count = count + 1
    print(count)
            
```

```python
check_auth(authors)
```
